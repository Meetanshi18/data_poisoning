```python
import random
```

```python
# Random generated data
loan_amnt = [] 
annual_inc = []
dti = []
delinq_2yrs = []
inq_last_6mths = []
pub_rec = []
open_acc = []
pub_rec_1 = []
revol_bal = []
revol_util = []
total_acc = []
collections_12_mths_ex_med = []
acc_now_delinq = []
emp_length = []
earliest_cr_line = []
mths_since_last_delinq = []
mths_since_last_record = []
mths_since_last_major_derog = []
term_36months = [] # 0 or 1
####################################
addr_state = []
purpose = []
verification_status = []
home_ownership = []

for _ in range(50000):
    loan_amnt.append(random.random())
    annual_inc.append(random.random()/10)
    dti.append(random.random())
    delinq_2yrs.append(0 if random.random() > 0.2 else 0.05)
    inq_last_6mths.append(0 if random.random() > 0.5 else random.random()/5)
    pub_rec.append(0 if random.random() > 0.2 else 0.002)
    open_acc.append(random.random()/5)
    pub_rec_1.append(0 if random.random() > 0.2 else 1)
    revol_bal.append(random.random()/100)
    revol_util.append(random.random()/5)
    total_acc.append(random.random()/5)
    collections_12_mths_ex_med.append(0 if random.random() > 0.2 else 0.1)
    acc_now_delinq.append(0 if random.random() > 0.2 else 0.1)
    emp_length.append(1 if random.random() > 0.5 else random.random())
    earliest_cr_line.append(random.random()+0.3)
    mths_since_last_delinq.append(1 if random.random() > 0.25 else random.random()/2)
    mths_since_last_record.append(1 if random.random() > 0.25 else random.random()/2)
    mths_since_last_major_derog.append(1 if random.random() > 0.25 else random.random()/2)
    term_36months.append(1 if random.random() > 0.4 else 0)
    addr_state.append(random.randint(1,50))
    purpose.append(1 if random.random() > 0.5 else (5 if random.random() > 0.5 else (13 if random.random() > 0.4 else random.randint(1,13))))
    home_ownership.append(4 if random.random() > 0.5 else (2 if random.random() > 0.5 else (1 if random.random() > 0.4 else random.randint(1,4))))
    verification_status.append(1 if random.random() > 0.25 else 0)
```

```python
import pandas as pd
df = pd.DataFrame({'loan_amnt': loan_amnt, 'annual_inc':annual_inc, 'dti':dti, 'delinq_2yrs':delinq_2yrs, 
                   'inq_last_6mths':inq_last_6mths, 'pub_rec':pub_rec, 'open_acc':open_acc, 'pub_rec_1':pub_rec_1, 
                   'revol_bal':revol_bal, 'revol_util':revol_util, 'total_acc':total_acc, 'collections_12_mths_ex_med':collections_12_mths_ex_med, 
                   'acc_now_delinq':acc_now_delinq, 'emp_length':emp_length, 'earliest_cr_line':earliest_cr_line, 
                   'mths_since_last_delinq':mths_since_last_delinq, 'mths_since_last_record':mths_since_last_record, 
                   'mths_since_last_major_derog':mths_since_last_major_derog, 'term_36months':term_36months, 
                   'addr_state':addr_state, 'purpose':purpose, 'home_ownership':home_ownership, 'verification_status':verification_status},                   
        columns=['loan_amnt', 'annual_inc', 'dti' ,'delinq_2yrs','inq_last_6mths','pub_rec','open_acc','pub_rec_1',
                 'revol_bal','revol_util','total_acc','collections_12_mths_ex_med','acc_now_delinq','emp_length','earliest_cr_line',
                 'mths_since_last_delinq','mths_since_last_record','mths_since_last_major_derog','term_36months','addr_state',
                 'purpose','home_ownership','verification_status'])
```

```python
# purpose: The purpose of the loan (takes values “credit_card”, “debt_consolidation”, “educational”, “major_purchase”, “small_business”, and “all_other”).
# int.rate: The interest rate of the loan, as a proportion (a rate of 11% would be stored as 0.11). Borrowers judged by LendingClub.com to be more risky are assigned higher interest rates.
# earliest_cr_line: The month the borrower's earliest reported credit line was opened
# emp_length Employment length in years. Possible values are between 0 and 10 where 0 means less than one year
# mths_since_last_delinq The number of months since the borrower's last delinquency.
# mths_since_last_major_derog Months since most recent 90-day or worse rating
# mths_since_last_record The number of months since the last public record.
# annual.inc: The self-reported annual income of the borrower.
# dti: The debt-to-income ratio of the borrower (amount of debt divided by annual income).
# revol.bal: The borrower’s revolving balance (amount unpaid at the end of the credit card billing cycle).
# total_acc The total number of credit lines currently in the borrower's credit file
# revol.util: The borrower’s revolving line utilization rate (the amount of the credit line used relative to total credit available).
# open_acc: The number of open credit lines in the borrower's credit file.
# inq.last.6mths: The borrower’s number of inquiries by creditors in the last 6 months.
# collections_12_mths_ex_med Number of collections in 12 months excluding medical collections
# delinq.2yrs: The number of times the borrower had been 30+ days past due on a payment in the past 2 years.
# pub.rec: The borrower’s number of derogatory public records (bankruptcy filings, tax liens, or judgments).
```
